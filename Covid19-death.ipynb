#### Dropping unnecessary columns

```python
death_1=death.drop(['Province/State','Lat','Long','Region Code','Region Name','Sub-region Code','Sub-region Name','Intermediate Region Code','Intermediate Region Name'],axis=1)
death_1=death_1.rename(columns={"ISO 3166-1 Alpha 3-Codes":"CODE"})
```

#### Renaming columns

```python
death_1=death_1.rename(columns={"Country/Region":"Country"})
```

```python
death_1.head(2)
```

#### Selecting required rows and columns

```python
sum_columns=list(death_1.select_dtypes(include=['float']).columns.values)

death_1=death_1.iloc[1:,:]
death_1.head()
```

### Data manipulation

#### Grouping data by Country and ISO3 Alpha code

```python
death_1=death_1.groupby(['Country','CODE'])[sum_columns].sum().reset_index()
#Creating a copy of dataframe for future usage
death_2=death_1.copy(deep=True)
death_1.shape
```

```python
death_2.shape
```

#### Calculating counts of confirmed cases day wise

```python
shape=list(death_1.shape)
y=shape[1]
y
```

```python
death_1.head(5)
```

```python
#loop function to calculate the deaths individual day wise

for i in range(shape[1]-1,2,-1):
    death_1.iloc[:,i] = death_1.iloc[:,i]-death_1.iloc[:,i-1]
```

### Data visualization

#### Generating World map for Date wise deaths

##### Extracting column names for Dates and Plot types

```python
y=list(death_1.shape)
shape1=y[1]

column_name = list(death_1.columns)
column_names = column_name[2:shape1]

plots=  ['equirectangular', 'mercator', 'orthographic', 'natural earth', 'kavrayskiy7', 'miller', 'robinson', 'eckert4',
            'azimuthal equal area', 'azimuthal equidistant', 'conic equal area', 'conic conformal', 'conic equidistant',
            'gnomonic', 'stereographic', 'mollweide', 'hammer',
            'transverse mercator', 'albers usa', 'winkel tripel',
            'aitoff', 'sinusoidal']
```

##### Generating World map for deaths date wise

```python
#plot function
def set_plot(headers_x,plotType):
    
    death_viz=death_1.copy(deep=True)

    death_viz['text'] = death_viz['Country'].astype(str) + '('  + 'code:' + death_viz['CODE'].astype(str)+ ')' + '<br>' \
                    'count: ' + death_viz[headers_x].astype(str)
    
    fig = px.choropleth(death_viz, locations=death_viz["CODE"],
                    color=death_viz[headers_x], # lifeExp is a column of gapminder
                    hover_name=death_viz["Country"], # column to add to hover information
                    color_continuous_scale=px.colors.sequential.Plasma)
        
    fig.update_layout(title_text='Country wise Total display count',
                geo=dict(
                showframe=True,
                showcoastlines=True,
                oceancolor="aqua",
                showocean=True,
                lakecolor="LightBlue",
                projection_type=plotType
                ))
            
    fig.show()

#define widgets
headers_x = widgets.Dropdown(
        options=column_names,
        value=column_names[0],
        description='Date'
    )

headers_x.set_title  = 'headers_x'

plotType = widgets.Dropdown(
        options=plots,
        value=plots[0],
        description='Plot Type'
    )

#interact function
interact(set_plot, headers_x = headers_x, plotType = plotType)
```

#### Visualisation of Death reported day wise country wise

##### Extracting Country list for filter purpose

```python
death_1.head()

country = list(death_1.iloc[:,0].unique())
```

##### Converting wide dataframe to long for data manipulation

```python
death_1_long = pd.DataFrame(columns=['Date','Country','Death'])

death_1_listdic=[]

for i in range(2,y[1]-1,1):
    for j in range(0,y[0],1):
        death_1_listdic.append({'Date':death_1.columns[i],'Country':death_1.at[j,'Country'],'Death':int(death_1.iloc[j,i])})
        
death_1_long=death_1_long.append(death_1_listdic)
death_1_long['Date']=pd.to_datetime(death_1_long['Date'])

death_2_long = pd.DataFrame(columns=['Date','Country','Death'])

death_2_listdic=[]

for i in range(2,y1[1],1):
    for j in range(0,y1[0],1):
        death_2_listdic.append({'Date':death_2.columns[i],'Country':death_2.at[j,'Country'],'Death':int(death_2.iloc[j,i])})
        
death_2_long=death_2_long.append(death_2_listdic)
death_2_long['Date']=pd.to_datetime(death_2_long['Date'])
```

##### Line chart Visualisation for Covid-19 deaths date wise Country wise (continuous and individual)

```python
chart_type=['Individual date','Continuous growth']

def line_chart(country_list,chart_type):
        
        if chart_type == 'Individual date':
            country_wise_ind=death_1_long.groupby(['Date','Country'])['Death'].sum().reset_index()
            country_wise_viz=country_wise_ind.loc[country_wise_ind['Country']==country_list]
            fig = px.line(country_wise_viz, x="Date", y="Death", title='Covid deaths date wise in '+country_list)
            fig.show()
            
        if chart_type == 'Continuous growth':
            country_wise_cont=death_2_long.groupby(['Date','Country'])['Death'].sum().reset_index()
            country_wise_viz=country_wise_cont.loc[country_wise_cont['Country']==country_list]
            fig = px.line(country_wise_viz, x="Date", y="Death", title='Covid deaths over time in '+country_list)
            fig.show()
        
#define widgets
country_list = widgets.Dropdown(
        options=country,
        value=country[0],
        description='Country'
    )

chart_type = widgets.Dropdown(
        options=chart_type,
        value=chart_type[0],
        description='Chart type'
    )

interact(line_chart,country_list=country_list,chart_type=chart_type)
```

##### Bar chart Visualisation for Covid-19 deaths date wise Country wise (continuous and individual)

```python
chart_type=['Individual date','Continuous growth']

def bar_chart(country_list,chart_type):
        
        if chart_type == 'Individual date':
            country_wise_ind=death_1_long.groupby(['Date','Country'])['Death'].sum().reset_index()
            country_wise_viz=country_wise_ind.loc[country_wise_ind['Country']==country_list]
            fig = px.bar(country_wise_viz, x="Date", y="Death", title='Covid deaths date wise in '+country_list)
            fig.add_trace(go.Scatter(x=country_wise_viz["Date"],y=country_wise_viz["Death"],name="trend"))
            fig.show()
            
        if chart_type == 'Continuous growth':
            country_wise_cont=death_2_long.groupby(['Date','Country'])['Death'].sum().reset_index()
            country_wise_viz=country_wise_cont.loc[country_wise_cont['Country']==country_list]
            fig = px.bar(country_wise_viz, x="Date", y="Death", title='Covid deaths over time in '+country_list)
            fig.add_trace(go.Scatter(x=country_wise_viz["Date"],y=country_wise_viz["Death"],name="trend"))
            fig.show()
        
#define widgets
country_list = widgets.Dropdown(
        options=country,
        value=country[0],
        description='Country'
    )

chart_type = widgets.Dropdown(
        options=chart_type,
        value=chart_type[0],
        description='Chart type'
    )

interact(bar_chart,country_list=country_list,chart_type=chart_type)
```

##### Bar chart Visualisation for top countries across dates

```python
chart_type=['Individual date','Continuous growth']

positions = [i for i in range(len(country))]

def bar_chart(chart_type,positions,Date):
        
        if chart_type == 'Individual date':
            country_wise_ind=death_1_long.groupby(['Date','Country'])['Death'].sum().reset_index().sort_values(by=['Death','Date'],ascending=False)
            country_wise_viz=country_wise_ind.loc[country_wise_ind['Date']==Date]
            country_wise_viz=country_wise_viz.head(positions)
            fig = px.bar(country_wise_viz, x="Country", y="Death", title="top "+str(positions)+" countries affected on "+Date+"("+chart_type+")")
            #fig.add_trace(go.Scatter(x=country_wise_viz["Date"],y=country_wise_viz["Confirmed"],name="trend"))
            fig.show()
            
        if chart_type == 'Continuous growth':
            country_wise_cont=death_2_long.groupby(['Date','Country'])['Death'].sum().reset_index().sort_values(by=['Death','Date'],ascending=False)
            country_wise_viz=country_wise_cont.loc[country_wise_cont['Date']==Date]
            country_wise_viz=country_wise_viz.head(positions)
            fig = px.bar(country_wise_viz, x="Country", y="Death", title="top "+str(positions)+" countries affected on "+Date+"("+chart_type+")")
            #fig.add_trace(go.Scatter(x=country_wise_viz["Date"],y=country_wise_viz["Confirmed"],name="trend"))
            fig.show()
        
#define widgets

chart_type = widgets.Dropdown(
        options=chart_type,
        value=chart_type[0],
        description='Chart type'
    )

Date = widgets.Dropdown(
        options=column_names_1,
        value=column_names_1[0],
        description='Date'
    )

Position = widgets.Dropdown(
        options=positions,
        value=positions[0],
        description='Position'
    )


interact(bar_chart,chart_type=chart_type,positions=Position,Date=Date)
```
